```python
import json
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Path to the JSON file
data = '/Users/sanjanamortha/Desktop/Desktop/Ecowear/Data_analysis/scores_2024604/1717200413849.json'
```

```python
# path to json file

with open(data, 'r') as file:
    data = json.load(file)
```

```python

# Extracting data
trial1_reaction_times = [response["reactionTime"] for response in data["FigureRecall"]["trialResponses"]["trial1"]]
trial2_reaction_times = [response["reactionTime"] for response in data["FigureRecall"]["trialResponses"]["trial2"]]
completion_times = data["FigureRecall"]["analysisResult"]["completionTimes"]
serial_position_effects_trial1 = [
    data["FigureRecall"]["serialPositionEffect"][0]["PrimacyEffect"],
    data["FigureRecall"]["serialPositionEffect"][0]["RecencyEffect"],
    data["FigureRecall"]["serialPositionEffect"][0]["MiddleEffect"]
]
serial_position_effects_trial2 = [
    data["FigureRecall"]["serialPositionEffect"][1]["PrimacyEffect"],
    data["FigureRecall"]["serialPositionEffect"][1]["RecencyEffect"],
    data["FigureRecall"]["serialPositionEffect"][1]["MiddleEffect"]
]

# Extracting item names
trial1_item_names = [response["selectedValue"] for response in data["FigureRecall"]["trialResponses"]["trial1"]]
trial2_item_names = [response["selectedValue"] for response in data["FigureRecall"]["trialResponses"]["trial2"]]
```

```python
# Verify the extracted data
print("Trial 1 Reaction Times:", trial1_reaction_times)
print("Trial 2 Reaction Times:", trial2_reaction_times)
print("Completion Times:", completion_times)
print("Serial Position Effects Trial 1:", serial_position_effects_trial1)
print("Serial Position Effects Trial 2:", serial_position_effects_trial2)
print("Trial 1 Item Names:", trial1_item_names)
print("Trial 2 Item Names:", trial2_item_names)
```

```python
# Plotting Reaction Times for each Trial
fig, axs = plt.subplots(3, 1, figsize=(10, 15))

# Reaction times
bar_width = 0.35
index = np.arange(len(trial1_reaction_times))

bars1 = axs[0].bar(index, trial1_reaction_times, bar_width, label='Trial 1')
bars2 = axs[0].bar(index + bar_width, trial2_reaction_times, bar_width, label='Trial 2')

axs[0].set_xlabel('Items')
axs[0].set_ylabel('Reaction Time (ms)')
axs[0].set_title('Reaction Times per Item')
axs[0].set_xticks(index + bar_width / 2)
axs[0].set_xticklabels(trial1_item_names, rotation=45)
axs[0].legend()
```

```python
# Completion times
axs[1].bar(['Trial 1', 'Trial 2'], completion_times, color=['blue', 'orange'])
axs[1].set_xlabel('Trials')
axs[1].set_ylabel('Completion Time (ms)')
axs[1].set_title('Completion Times per Trial')

# Serial Position Effects
index = np.arange(3)
bar_width = 0.35

bars1 = axs[2].bar(index, serial_position_effects_trial1, bar_width, label='Trial 1')
bars2 = axs[2].bar(index + bar_width, serial_position_effects_trial2, bar_width, label='Trial 2')

axs[2].set_xlabel('Serial Position Effect')
axs[2].set_ylabel('Effect Value')
axs[2].set_title('Serial Position Effects per Trial')
axs[2].set_xticks(index + bar_width / 2)
axs[2].set_xticklabels(['Primacy', 'Recency', 'Middle'])
axs[2].legend()

plt.tight_layout()
plt.show()
```
